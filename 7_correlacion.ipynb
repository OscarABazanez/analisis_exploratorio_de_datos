# CorrelaciÃ³n

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()
data = breast_cancer.data
features = breast_cancer.feature_names
df = pd.DataFrame(data, columns = features)
df
```

```
df_small = df.iloc[:,:6]
correlation_mat = df_small.corr()
sns.heatmap(correlation_mat, annot = True)
plt.show()
```

Convirtiendo a vector

```
corr_values=correlation_mat.unstack()
corr_values
```

Escogemos el tipo de algorimot para ordenar nuestro vector.

```
sorted_pairs = corr_values.sort_values(kind="quicksort")
```

## Graficando la CorrelaciÃ³n con seaborn

```
sns.pairplot(df_small)
```

## Graficando la CorrelaciÃ³n con matplotlib

```
negative_pairs = sorted_pairs[sorted_pairs < 0]
corr = df_small.corr()
sns.set()
cols = ['worst texture', 'concave points error', 
        'perimeter error',
        'mean compactness', 
        'worst symmetry',
        'mean radius', 
        'mean area']
sns.pairplot(df[cols], size = 2.5)
plt.show();
```
